# Problem statement

This not only jeopardizes customer satisfaction but also impacts the company's reputation and market competitiveness.

**5.Lack of Predictive Maintenance System:** The core of the problem lies in the absence of a comprehensive predictive maintenance system. The company currently lacks the capability to anticipate and address potential machinery failures before they lead to critical breakdowns. This deficiency hampers the overall efficiency of the manufacturing process.

## Resampling

```python
##Importing resample from *sklearn.utils* package.
from sklearn.utils import resample

# Separate the case of 0 and 1
no_failure_0 = df[df.Target==0]
failure_1 = df[df.Target==1]

#Upsample the failuer(1) cases.
df_minority_upsampled = resample(failure_1,replace=True,n_samples=2000)

# Combine majority class with upsampled minority class
new_df = pd.concat([no_failure_0, df_minority_upsampled])
```

## Shuffle the data

```python
from sklearn.utils import shuffle
shuffle_df = shuffle(new_df)
```

```python
encodeded_df = pd.get_dummies(shuffle_df.drop(["Product ID",'Target', 'Failure Type'],axis=1),drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(encodeded_df)
```

```python
scaled
```

```python
# Split the dataset into features and targets (y)
y = shuffle_df['Target']  # Target variable (failure or not)
```

## Binary classification

### In this dataset there are two target variables

### 1) Failure or Non Failure
### 2) Types of Failure

### First, we are going with Failure and Non Failure classification

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(scaled, y, test_size=0.2, random_state=42)
```

### Configure the AutoML model

```python
# Initialize and configure the AutoML model
automl = AutoML(algorithms=["Decision Tree", "Random Forest", "Extra Trees"],
                total_time_limit=60,  # Maximum time in seconds for training
                mode="Explain",
                eval_metric="accuracy",  # Evaluation metric
                optuna_time_budget=300,  # Maximum time in seconds for hyperparameter optimization
                optuna_verbose=True)  # Verbose output for hyperparameter optimization
```

```python
warnings.filterwarnings("ignore", message="findfont: Font family 'Arial' not found")
```

```python
plt.rcParams['font.family'] = 'sans-serif'
```

### Train the AutoML model for binary classification

```python
# Train the AutoML model
automl.fit(X_train, y_train)

```

```python
# Evaluate the trained model
score = automl.score(X_test, y_test)
print("Accuracy:", score)
```

```python
# Make predictions
predictions = automl.predict(X_test)
predictions
```

### Performance Report For Binary Classification

```python
performance = automl.report()
performance
```

## Multi-class classification

### Now we are going to classify types of failure

### First of all, we need to convert categorical Failure Type variable into numerical variable

### Evaluate the trained model

```python
from sklearn.preprocessing import LabelEncoder
y = shuffle_df['Failure Type']

# Encode the target variable 'y' using LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

### Split data into training and testing sets

```python
# Split data into training and testing sets
X_train, X_test, y_train_target, y_test_target = train_test_split(scaled, y_encoded, test_size=0.2, random_state=42)

```

### Configure AutoML Model

```python
# Initialize and configure the AutoML model
automl1 = AutoML(algorithms=["Decision Tree", "Random Forest", "Extra Trees"],
                total_time_limit=60,  # Maximum time in seconds for training
                mode="Explain",
                eval_metric="accuracy",  # Evaluation metric
                optuna_time_budget=300,  # Maximum time in seconds for hyperparameter optimization
                optuna_verbose=True)  # Verbose output for hyperparameter optimization
```

### Train the AutoML Model for multi-class classification

```python
# Train the AutoML model
automl1.fit(X_train, y_train_target, sample_weight=None, cv=None)
```

### Evaluate the train model

```python
# Evaluate the trained model
score_target1 = automl1.score(X_test, y_test_target)
print("Accuracy (Target):", score_target1)
```

```python
# Make predictions
predictions_target1 = automl1.predict(X_test)
predictions_target1
```
